```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\dap20\Downloads\customer_shopping_behavior.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
df.columns = (
    df.columns
    .str.replace('_', ' ', regex=False)   # first remove the weird underscores
    .str.split()                          # split on spaces
    .str.join('_')                        # join words with single underscore
    .str.lower()
)

df.columns
```

```python
#create a column age_group
labels = ['Young Adult', 'Adult','Middle-aged','Senior']
df['age_group'] = pd.qcut(df['age'],q=4, labels = labels)
```

```python
df[['age','age_group']].head(10)
```

```python
#create column purchase_frequency_days

frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annualy': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequencyofpurchases'].map(frequency_mapping)

df[['frequencyofpurchases', 'purchase_frequency_days']].head(10)
```

```python
df[['discountapplied','promocodeused']].head(10)
```
